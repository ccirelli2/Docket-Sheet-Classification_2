```python
# Module:  Basic Functions

def transform_ones(x):
    x = 1
    return x

def get_dict_freq_dist(Dataframe):
    Dict = {}
    for x in Dataframe['Time Period']:
        Dict[x] = Dict.get(x,0) + 1
    return Dict

def get_dict_conditional_probs(dict_TP_freq_dist, length_df):
    # Create new dict object to capture the conditional probabiliteis for each time period
    Dict_conditional_probs = {}
    
    # Get Keys for the TP freq dist
    for x in dict_TP_freq_dist:
        # Create a new dict whos keys are the TP's and values the freq / total lenght
        Dict_conditional_probs[x] = round(((dict_TP_freq_dist[x] / length_df)*100),2)
    
    return Dict_conditional_probs
    
```

```python
# Create Conditional Probability Table for Each Time Period

def get_conditional_prob(dataframe):
    
    # Get Dictionary Conditional Probabilities
    Dict_conditional_probs = {}
    
    # Get length of our dataframe. 
    List = [x for x in dataframe['Time Period']]
    get_ones = map(transform_ones, List)
    length_df = functools.reduce((lambda x,y: x+y), get_ones)
    
    # Get Time Period Freq Dist
    dict_time_period_freq_dist = get_dict_freq_dist(dataframe)

    # Get Conditional Probability for each time period
    dict_conditional_probability_distribution_Timeperiods = get_dict_conditional_probs(dict_time_period_freq_dist, length_df)
    
    # Return Conditional Probability Object
    return dict_conditional_probability_distribution_Timeperiods
```
