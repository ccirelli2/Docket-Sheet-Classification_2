```python
def get_collocation_stg8_stg11(df1):
    
    # Create a list to capture the value of 1 (match) or 0 (no match) for each case in 
    # the case_level_prediction file 
    List_capture_value = []
    
    # Create a list to capture the stage values prior to stage 11
    List = []
    
    # For each case number in the case_level_predictions file
    for x,y in zip(df1.itertuples(),df1[1:].itertuples()):        
        
        if y[1] == x[1]:
                
            if y[-1] < 11:
                List.append(y[-1])
            else:
                if 8 in List:
                    List_capture_value.append((y[1],1))
                else:
                    List_capture_value.append((y[1],0))   
                List = []
                
    return List_capture_value

     
        
```
